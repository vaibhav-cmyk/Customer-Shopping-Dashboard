```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\vaib9\OneDrive\Desktop\customer_shopping_behavior\customer_shopping_behavior.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ',"_")
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df.columns
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
labels = ['young','adult','middle_aged','senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
frequencies_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequencies_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(20)
```
